```python
import tensorflow as tf
from tensorflow import keras

import pandas as pd
import numpy as np

training_df = pd.read_csv('./data/train.csv')
color_dict = {'red': 0, 'blue': 1, 'green': 2, 'teal': 3, 'orange': 4, 'purple': 5}
training_df['color'] = training_df.color.apply(lambda x: color_dict[x])
np.random.shuffle(training_df.values)

training_df.head()
```

```python
test_df = pd.read_csv('./data/test.csv')
test_x = np.column_stack((test_df.x.values, test_df.y.values))

model = keras.Sequential([
    keras.layers.Dense(32, input_shape=(2,), activation="relu"),
    keras.layers.Dense(32, activation="relu"),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(6, activation="sigmoid")]) # There are 6 colors

model.compile(optimizer="adam", loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
             metrics=['accuracy'])

x = np.column_stack((training_df.x.values, training_df.y.values))

model.fit(x, training_df.color.values, batch_size=2, epochs=10)

print("\nEVALUATION")
test_df['color'] = test_df.color.apply(lambda x: color_dict[x])
model.evaluate(test_x, test_df.color.values)

print("\nColors:     [r  b  g  t  o  p]")
print("Prediction", np.round(model.predict(np.array([[0,1]]))))
print("Prediction", np.round(model.predict(np.array([[0,3]]))))
print("Prediction", np.round(model.predict(np.array([[0,5]]))))
print("Prediction", np.round(model.predict(np.array([[-2,1]]))))
print("Prediction", np.round(model.predict(np.array([[-2,3]]))))
print("Prediction", np.round(model.predict(np.array([[-2,5]]))))
```
