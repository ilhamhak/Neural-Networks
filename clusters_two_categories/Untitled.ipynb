```python
from tensorflow import keras

import pandas as pd
import numpy as np

training_df = pd.read_csv('./data/train.csv')
cat_color = pd.get_dummies(training_df.color).values
cat_marker = pd.get_dummies(training_df.marker).values
labels = np.concatenate((cat_color, cat_marker), axis=1)
x = np.column_stack((training_df.x.values, training_df.y.values))


test_df = pd.read_csv('./data/test.csv')
test_cat_color = pd.get_dummies(test_df.color).values
test_cat_marker = pd.get_dummies(test_df.marker).values
test_labels = np.concatenate((test_cat_color, test_cat_marker), axis=1)
test_x = np.column_stack((test_df.x.values, test_df.y.values))

np.random.RandomState(seed=27).shuffle(x)
np.random.RandomState(seed=27).shuffle(labels)

model = keras.Sequential([
    keras.layers.Dense(81, input_shape=(2,), activation="relu"),
    keras.layers.Dropout(0.42),
    keras.layers.Dense(81, activation="relu"),
    keras.layers.Dropout(0.42),
    keras.layers.Dense(9, activation="sigmoid")]) # There are 6 colors 3 shapes

model.compile(optimizer="adam", loss=keras.losses.BinaryCrossentropy(),
             metrics=['accuracy'])

model.fit(x, labels, batch_size=9, epochs=20)

print("\nEVALUATION")
model.evaluate(test_x, test_labels)

print("\nColor&Shape:[r  b  g  t  o  p  ^  +  *]")
print("Prediction", np.round(model.predict(np.array([[0,1]]))))
print("Prediction", np.round(model.predict(np.array([[0,3]]))))
print("Prediction", np.round(model.predict(np.array([[0,5]]))))
print("Prediction", np.round(model.predict(np.array([[-2,1]]))))
print("Prediction", np.round(model.predict(np.array([[-2,3]]))))
print("Prediction", np.round(model.predict(np.array([[-2,5]]))))
```
